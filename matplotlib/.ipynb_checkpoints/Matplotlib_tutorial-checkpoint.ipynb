```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
# 256 points from negative pi to pi, including the endpoint pi
C, S = np.cos(X), np.sin(X)
# Arrays of Sin and Cosine operators on X

for styleitem in plt.style.available:
    print styleitem
    plt.style.use(styleitem)
    plt.plot(X, C, 'c--', linewidth=3, label='Cosine')
    plt.plot(X, S, 'r--', linewidth=3, label='Sin')
    plt.legend()
    plt.axis([-np.pi, np.pi, -1.2, 1.2])
    plt.show()
```

```python
t = 2 * np.pi / 3

# Trick to plot a straight vertical line
plt.plot([t, t], [0, np.cos(t)], color='blue', linewidth=2.5, linestyle="--")

# Plots the point on the cosine line
plt.scatter([t, ], [np.cos(t), ], 50, color='blue')


plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
            xy=(t, np.cos(t)), xycoords='data',
            xytext=(-90, -50), textcoords='offset points',
            fontsize=16,
            arrowprops=dict(arrowstyle="->",
                            connectionstyle="arc3,rad=.2"))

#plot red line to origin
plt.plot([t, t], [0, np.sin(t)], color='red', linewidth=2.5, linestyle="--")
plt.scatter([t, ], [np.sin(t), ], 50, color='red')

plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
            xy=(t, np.sin(t)), xycoords='data',
            xytext=(+10, +30), textcoords='offset points',
            fontsize=16,
            arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
```

```python
n = 256
X = np.linspace(-np.pi, np.pi, n, endpoint=True)
Y = np.sin(2 * X)

plt.style.use("seaborn-white")

plt.plot(X, Y + 1, color='blue', alpha=.50)
plt.fill_between(X, Y + 1, 1, alpha=0.5)
plt.plot(X, Y - 1, color='blue', alpha=1.00)
plt.fill_between(X, -1, Y - 1, where=(Y-1>-1), color='blue', alpha=0.5)
plt.fill_between(X, Y - 1, -1, where=(Y-1<-1), color='red', alpha=0.5)
plt.show()
```

```python
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
T = np.arctan2(Y, X)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.scatter(X, Y, s=75, c=T, alpha=.5)

plt.set_cmap('jet')

plt.xlim(-1.5, 1.5)
plt.xticks(())
plt.ylim(-1.5, 1.5)
plt.yticks(())
plt.show()
```

```python
n = 12
X = np.arange(n)
Y1 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)
Y2 = (1 - X / float(n)) * np.random.uniform(0.5, 1.0, n)

plt.bar(X, +Y1, facecolor='#9999ff', edgecolor='white')

for x, y in zip(X, Y1):
    plt.text(x + 0.4, y + 0.05, '%.2f' % y, ha='center', va='bottom')
    

plt.bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x, y in zip(X, -Y2):
    plt.text(x + 0.4, y - 0.05, '%.2f' % y, ha='center', va='top')
    

plt.ylim(-1.25, +1.25)
plt.show()
```

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 256
x = np.linspace(-3, 3, n)
y = np.linspace(-3, 3, n)
X, Y = np.meshgrid(x, y)

plt.axes([0.025, 0.025, 0.95, 0.95])

plt.contourf(X, Y, f(X, Y), 8, alpha=.75, cmap='hot')
C = plt.contour(X, Y, f(X, Y), 8, colors='black', linewidth=.5)
plt.clabel(C, inline=1, fontsize=10)


plt.xticks(())
plt.yticks(())

plt.show()
```

```python
def f(x, y):
    return (1 - x / 2 + x ** 5 + y ** 3) * np.exp(-x ** 2 - y ** 2)

n = 10
x = np.linspace(-3, 3, 4 * n)
y = np.linspace(-3, 3, 3 * n)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.axes([0.025, 0.025, 0.95, 0.95])
plt.imshow(Z, interpolation='nearest', cmap='bone', origin='lower')
plt.colorbar(shrink=.92)

plt.xticks(())
plt.yticks(())

plt.show()
```
